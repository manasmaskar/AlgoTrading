```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
bre_nt_df = pd.read_csv('cointegrated_pairs_data/BRENT_df.csv')
light_df = pd.read_csv('cointegrated_pairs_data/LIGHT_df.csv')
```

```python
# Plot both time series
plt.figure(figsize=(10, 6))
plt.plot(bre_nt_df['Time'], bre_nt_df['Close'], label='BRENT_df')
plt.plot(light_df['Time'], light_df['Close'], label='LIGHT_df')
plt.title('Time Series for BRENT and LIGHT')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import coint

# Run the Engle-Granger Cointegration Test
score, p_value, _ = coint(bre_nt_df['Close'], light_df['Close'])
print(f"Cointegration p-value: {p_value:.4f}")
print(score)
```

<h3>Interpretation --> <br>
<h6>The p-value is extremely low, indicating strong evidence of cointegration between the BRENT_df and LIGHT_df time series. This suggests that these two time series share a long-term equilibrium relationship, meaning their price movements are related in the long run.
