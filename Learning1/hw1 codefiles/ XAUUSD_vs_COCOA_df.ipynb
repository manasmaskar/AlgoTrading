```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
XAUUSD_df = pd.read_csv('/Users/manasmaskar/Rutgers/Fall24/Algo Trading/HomeWork/HW1/hw1 codefiles/cointegrated_pairs_data/XAUUSD_Candlestick_1_Hour_BID_01_df.csv')
COCOA_df = pd.read_csv('/Users/manasmaskar/Rutgers/Fall24/Algo Trading/HomeWork/HW1/hw1 codefiles/cointegrated_pairs_data/COCOA_df.csv')
```

```python
# Plot both time series
plt.figure(figsize=(10, 6))
plt.plot(XAUUSD_df['Time'], XAUUSD_df['Close'], label='XAUUSD_df')
plt.plot(COCOA_df['Time'], COCOA_df['Close'], label='COCOA_df')
plt.title('Time Series for XAUUSD and COCOA')
plt.legend()
plt.show()
```

<h3>Interpretation --><br>


<h6>This kind of chart might be useful for comparing the stability of a safe-haven asset (gold) versus a more volatile commodity (cocoa). <br>
Investors might use this to assess risk: while gold shows steady, lower-risk returns, cocoa offers potential for higher gains but with greater volatility.

```python
from statsmodels.tsa.stattools import coint

# Run the Engle-Granger Cointegration Test
score, p_value, _ = coint(XAUUSD_df['Close'], COCOA_df['Close'])
print(f"Cointegration p-value: {p_value:.4f}")
print(score)
```
